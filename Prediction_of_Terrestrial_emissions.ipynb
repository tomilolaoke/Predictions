#Introduction:

The United Nations and environmental studies have proven that Greenhouse gases are major contributors to climate change and its impacts. Climate change is responsible for weather flunctuations and global temperature rise. Climate change can be linked to weather flunctuations and biodiversity loss and other environmental impacts vital to human survival and existence. This makes it imperative to apply technology in forcasting greenhouse emission as an effort to mitigate the negative impacts of climate change, adapt and mitigate its inevitable effects and enable regulatory bodies make informed decisions on how to reduce emissions in the future.  

Data Source:
The dataset was majorly created by the UK National Atmospheric Emmissions Inventory and the DESNZ National Statistics of energy consumption for Local Authority areas. It includes details on emissions of greegnhouse gases:methane, nitrous oxide and carbon dioxide. The dataset covers territorial carbon dioxide emissions within UK borders only. It focuses on emission generation from various sectors and sub-sectors excluding the transportation, aviation and military and  cuts across over 300  Local areas.[1]

The data sources include:Emissions from large installations, UK centre for Ecology and Hydrology,Emission distribution maps from NAEI programme. The data set includes 15 variables and 523643 observations

This study aims at analysing the use of supervised learning models such as Multiple linear regression and Least Absolute Shrinkage and Selection Operation (LASSO) in prediction territorial emissions.

Motivation for the use of AI: Machine learning algorithms are effective in detecting trends and insights that can be usefule in decision making. Particularly for time series datasets,they can be applied to evaluate accurate forcast of weather,

Evaluation Metrics:


Mean squared error:

$MSE=\frac{1}{n}\sum_{i=1}^{n}(y_i-\hat{y}_i)^2$, where:

n is the number observations


$y_i$ is the actual of y(predictor-Territorial emissions (kt CO2e))


$\hat{y}_i$ is the predicted values of yCoefficient of determination($R_2$):

Other metrics are:

Mean Absolute Error:$\frac{1}{n}\sum_{i=1}^{n}|y_i-\hat{y}_i|^2$,

Root Mean Square Error=$\sqrt{MSE}$


coefficient of determination($R^2):1-\frac{\sum_{i=1}^{n}(y_i-\hat{y}_i)^2}{\sum_{i=1}^{n}(y_i-\bar{y}_i)^2}$


```python
#estimate coefficient of determination on test
print(fit_model.score(x_test,y_test))
```
